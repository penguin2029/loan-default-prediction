```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/loan_data.csv")
```

```python
df.drop(columns = ['Unnamed:0'] , inplace=True , errors = 'ignore')
```

```python
print('before cleaning :',df.shape)
```

```python
print(df.columns.tolist())
```

```python
df = df.drop(columns=["Unnamed: 0"])
print(df.shape)
print(df.columns.tolist())
```

TASK1 - REMOVE BAD AGE ROWS

```python
df = df[(df['age'] > 0) & (df['age'] <= 100)]
print('after removing the bad age rows:',df.shape)
```

df["age"] > 0       → keep rows where age is more than 0
df["age"] <= 100    → keep rows where age is 100 or less
&                   → BOTH conditions must be true
df = df[...]        → overwrite df with only the rows that passed

TASK2 - FILL MISSING MONTHLY INCOME

```python
median_income = df['MonthlyIncome'].median()
print("median_income:",median_income)

df['MonthlyIncome'] = df['MonthlyIncome'].fillna(median_income)
print("Missing values in MonthlyIncome:", df["MonthlyIncome"].isnull().sum())
```

df["MonthlyIncome"].median()   → calculate the middle value of income column
.fillna(median_income)         → wherever there is a NaN, put median_income there
df["MonthlyIncome"] = ...      → save the result back into the same column
.isnull().sum()                → verify — should now show 0 missing

TASK3 - FILL MISSING NUMBER OF DEPENDENTS

```python
median_dependants = df['NumberOfDependents'].median()
print('median_dependants',median_dependants)

df['NumberOfDependents'] = df['NumberOfDependents'].fillna(median_dependants)
print('Missing values in no. of dependants:',df['NumberOfDependents'].isnull().sum())
```

.median()        → finds middle value of dependents column
.fillna(...)     → fills all empty cells with that median value
.isnull().sum()  → should show 0 after filling

Median dependents is 0 — meaning more than half the borrowers in this dataset have no dependents.

TASK4 - HANDLE INCOME OUTLIERS

```python
df['MonthlyIncome'] = df['MonthlyIncome'].clip(upper=50000)

print('max value after clipping',df['MonthlyIncome'].max())
print('mean value after clipping',df['MonthlyIncome'].mean().round(2))
```

.clip(upper=50000)   → any value above 50000 becomes 50000
                       values below 50000 stay exactly as they are

TASK 5 - VERIFY EVERYTHING IS CLEAN

print('Final shape:',  df.shape)
print('Missing values:',df.isnull().sum())
print('\nAge min',  df['age'].min())
print('Age max',  df['age'].max())
print('max monthly income',  df['MonthlyIncome'].max())

```python
TASK6 - SAVE THE CLEAN DATA
```

```python
df.to_csv("data/loan_data_cleaned.csv",index=False)
print('clean data saved successfully!')
```

df.to_csv(...)        → saves df as a CSV file
"data/loan_..."       → saves it inside your data folder
index=False           → don't save the row numbers as a column
                        remember the Unnamed: 0 problem? this prevents that
